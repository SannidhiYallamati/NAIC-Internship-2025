```python
# Task 1 ‚Äì Exploratory Data Analysis (EDA)
# Internship: NAIC Data Science Internship 2025
# Dataset: college_student_placement_dataset.csv

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("D:/internship/archive (5)/college_student_placement_dataset.csv")
print("‚úÖ Dataset loaded successfully!\n")

# Show dataset shape and columns
print("üìè Shape:", df.shape)
print("\nüßæ Column names:")
print(df.columns.tolist())

# Display first 5 rows
print("\nüîç Preview of dataset:")
display(df.head())

# Data types and non-null counts
print("\nüìÑ Dataset Info:")
df.info()

# Check for missing values
print("\n‚ùó Missing values per column:")
print(df.isnull().sum())

# Summary statistics
print("\nüìä Summary Statistics:")
print(df.describe(include='all'))

# üîç Step: Check unique values in each column
print("\nüîç Unique values in each column (to detect placement column):")
for col in df.columns:
    print(f"{col}: {df[col].unique()[:10]}")  # show only first 10 unique values

# Manually define placement status column if known
# You can change this name based on what print(df.columns) showed
status_col = 'status'  # ‚Üê change if it's different like 'Placed' or 'Placement_Status'

if status_col in df.columns:
    # Placement status count
    sns.countplot(x=status_col, data=df, palette='Set2')
    plt.title("Placement Status Count")
    plt.show()

    # Gender vs Placement
    sns.countplot(x='gender', hue=status_col, data=df, palette='coolwarm')
    plt.title("Placement by Gender")
    plt.show()

    # Specialisation vs Placement
    plt.figure(figsize=(10, 5))
    sns.countplot(x='specialisation', hue=status_col, data=df, palette='magma')
    plt.title("Specialisation vs Placement")
    plt.xticks(rotation=30)
    plt.show()

    # Degree percentage vs Placement
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x='degree_p', hue=status_col, kde=True)
    plt.title("Degree Percentage vs Placement")
    plt.show()

    # E-test score vs Placement
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x='etest_p', hue=status_col, kde=True)
    plt.title("E-test Score vs Placement")
    plt.show()
else:
    print(f"\n‚ö†Ô∏è Column '{status_col}' not found in dataset. Skipping placement visualizations.")

# Work experience
if 'workex' in df.columns:
    sns.countplot(x='workex', data=df, palette='Accent')
    plt.title("Work Experience Distribution")
    plt.show()

# Correlation heatmap (numeric columns only)
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlGnBu')
plt.title("Correlation Heatmap")
plt.show()

# ‚úÖ Summary
print("\n‚úÖ Task 1 ‚Äì EDA Completed.")
print("\nüìå Suggested Insights:")
print("1. Degree percentage and MBA scores impact placement chances.")
print("2. Gender and specialisation also influence placement trends.")
print("3. Candidates with work experience may have higher placement rates.")
```
