```python
# Task 3 ‚Äì Dashboard (Data Visualization)
# Internship: NAIC Data Science Internship 2025
# Dataset: college_student_placement_dataset.csv

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ---------------------------
# Step 1: Load Dataset
# ---------------------------
df = pd.read_csv("D:/internship/archive (5)/college_student_placement_dataset.csv")
print("‚úÖ Dataset Loaded Successfully!\n")

# Show basic info
print("üìå Available Columns:", df.columns.tolist())
print("üìè Shape:", df.shape)

# ---------------------------
# Step 2: Placement Status Count
# ---------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="Placement", hue="Placement", data=df, palette="Set2", legend=False)
plt.title("Overall Placement Status")
plt.show()

# ---------------------------
# Step 3: Placement by Internship Experience
# ---------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="Internship_Experience", hue="Placement", data=df, palette="coolwarm")
plt.title("Internship Experience vs Placement")
plt.show()

# ---------------------------
# Step 4: Placement by Academic Performance
# ---------------------------
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="Academic_Performance", hue="Placement", multiple="stack", palette="Set2")
plt.title("Academic Performance vs Placement")
plt.show()

# ---------------------------
# Step 5: CGPA vs IQ Scatter Plot
# ---------------------------
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="CGPA", y="IQ", hue="Placement", palette="magma")
plt.title("CGPA vs IQ by Placement")
plt.show()

# ---------------------------
# Step 6: Correlation Heatmap
# ---------------------------
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()

print("\n‚úÖ Task 3 ‚Äì Dashboard Visualization Completed Successfully!")
```
