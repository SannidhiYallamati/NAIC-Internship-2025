```python
# Task 3 â€“ Dashboard (Jupyter Notebook Version)
# Internship: NAIC Data Science Internship 2025
# Dataset: college_student_placement_dataset.csv

import pandas as pd
import plotly.express as px

# ---------------------------
# Step 1: Load Dataset
# ---------------------------
df = pd.read_csv("D:/internship/archive (5)/college_student_placement_dataset.csv")
print("âœ… Dataset Loaded")

# ---------------------------
# Step 2: Placement Status Count
# ---------------------------
fig_status = px.histogram(df, x="Placement", color="Placement",
                          title="Overall Placement Status",
                          barmode="group")
fig_status.show()

# ---------------------------
# Step 3: Placement by Internship Experience
# ---------------------------
fig_internship = px.histogram(df, x="Internship_Experience", color="Placement",
                              title="Internship Experience vs Placement",
                              barmode="group")
fig_internship.show()

# ---------------------------
# Step 4: Placement by Academic Performance
# ---------------------------
fig_academic = px.histogram(df, x="Academic_Performance", color="Placement",
                            title="Academic Performance vs Placement",
                            barmode="group")
fig_academic.show()

# ---------------------------
# Step 5: CGPA vs IQ Scatter Plot
# ---------------------------
fig_cgpa = px.scatter(df, x="CGPA", y="IQ", color="Placement",
                      title="CGPA vs IQ by Placement",
                      hover_data=["College_ID"])
fig_cgpa.show()
```
